```
import numpy as np
import matplotlib as plt
import pandas as pd
import seaborn as sns
```

```
tips = sns.load_dataset('tips')
tips.head(1)
```

```
sns.relplot(x='total_bill',y='tip',data=tips)
```

```
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',data=tips)
```

```
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',col='sex',data=tips) #col='day'
```

```
sns.catplot(x='day',y='total_bill',data=tips)
```

```
# sns.catplot(x='day',y='total_bill',data=tips,kind='swarm')
# sns.catplot(x='day',y='total_bill',data=tips,hue='time')
# sns.catplot(x='day',y='total_bill',data=tips,kind='box')
# sns.catplot(x='day',y='total_bill',data=tips,kind='boxen')
# sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
# sns.catplot(x='day',y='total_bill',data=tips,kind='box',hue='smoker')
# sns.catplot(x='day',y='tip',data=tips,kind='violin',hue='smoker',split=True)
sns.catplot(x='day',y='total_bill',data=tips,kind='bar',hue='sex')
```

```
titanic = sns.load_dataset('titanic')
titanic.head(1)
```

```
sns.catplot(x='sex',y='survived',hue='class',data=titanic,kind='bar')
```

```
sns.catplot(x='survived',data=titanic,kind='count')
```

```
sns.catplot(x='class',y='fare',hue='survived',data=titanic,kind='bar')
```

```
sns.regplot(x='total_bill',y='tip',data=tips)
```

```
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```
sns.lmplot(x='size',y='tip',data=tips,x_estimator=np.mean)
```

```
sns.lmplot(x='age',y='fare',data=titanic)
```

```
sns.lmplot(x='pclass',y='fare',data=titanic)
```

```
anscombe = sns.load_dataset('anscombe')
anscombe
```

```
sns.lmplot(x='x',y='y',col='dataset',hue='dataset',data=anscombe)
```

```
sns.lmplot(x='x',y='y',data=anscombe.query("dataset=='II'"),ci=None,order=2)
```

```
sns.lmplot(x='x',y='y',data=anscombe.query("dataset=='III'"),ci=None,robust=True)
```
